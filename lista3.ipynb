## Questão nº3:
## Escolha uma coluna com valores numéricos e apague todas as linhas que possuam valor nulo

```python
nova_data = data.dropna(axis=0, subset=["alcohol"])
```

## Questão n°4:
## Calcule a média desta mesma coluna

```python
media1 = nova_data["alcohol"].mean()
print(media1)
```

## Questão nº5:
## Escolha um critério para eliminar as linhas que possuam valores muito distantes da média e crie um novo Dataframe.

```python
# usarei padrão do valor ser 2 maior ou 2 menor
for j, n in enumerate(nova_data.alcohol):
  if (n < (media1 + 2) and n > (media1 - 2)):
    nova_data.drop(labels=[j], axis = 0, inplace = True)
  
nova_data_2 = nova_data
media2 = nova_data_2["alcohol"].mean()
print(media2)
```

## Questão n°6:
## Crie um novo Dataframe que contenha apenas 5 colunas do conjunto anterior.

```python
nova_data_3 = nova_data_2[['fixed acidity', 'quality', 'citric acid', 'pH', 'alcohol']]
  
print(nova_data_3)
```

## Questão n°7:
## Escolha duas colunas numéricas e crie uma nova coluna que siga a regra: (Coluna A - Coluna B)*número aleatório (entre 0 e 1).

```python
nova_data_3.insert(1, 'Ph_x_Alcohol', ((nova_data_3['pH'] + nova_data_3['alcohol']) * random.random()).astype(float))
print(nova_data_3)
display(nova_data_3)
```

## Questão n°8:
## Embaralhe aleatoriamente o dataframe (índices e valores).

```python
nova_data_4 = nova_data_3.reindex(np.random.permutation(nova_data_3.index))
print(nova_data_4)
display(nova_data_4)
```
